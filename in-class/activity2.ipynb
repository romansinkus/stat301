# A: Is the relationship between WAGE and EDUCATION statistically significant?

```r
wage_model <- lm(wage ~ education, data = wage_data)
get_regression_table(wage_model)
```

The p-value for education which is the B1_hat value is 0. Therefore, we can conclude that the relationship between wage and education is statistically significant since this value is below 0.05 (the significance level).
