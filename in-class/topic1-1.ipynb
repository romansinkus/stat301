```r
library(tidyverse)
library(infer)
library(cowplot)
library(broom)
library(moderndive)
```

```r
wage_data <- read.table("data/wage.txt", header=TRUE)
head(wage_data)
```

# (A): Plot WAGE versus EDUCATION. Is there a linear relationship between them?

```r
wage_scatterplot <- ggplot(wage_data, aes(education, wage)) +
  geom_point() +
  xlab("Education") +
  ylab("Wage") +
  theme(
    text = element_text(size = 16.5),
    plot.title = element_text(face = "bold"),
    axis.title = element_text(face = "bold")
  )
wage_scatterplot
```

There is a rough linear relationship between these two variables.

# (B): Find the correlation coefficient of the two variables. Is it consistent with what you observe in (A)?

```r
wage_data %>% 
  get_correlation(formula = education ~ wage)
```
