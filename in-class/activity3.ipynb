# Part A

```r
# Linear model before adjusting for sex
wage_model <- lm(wage ~ education, data = wage_data)
get_regression_table(wage_model)
```

```r
wage_model <- lm(wage ~ education + sex, data = wage_data)
get_regression_table(wage_model)
```

"Adjusting for SEX" is the same thing as creating an additive model. Therefore, we fit the model using `wage ~ education + sex`. In other words, we are creating a linear model such that `y = beta_0 + beta_1*EDUCATION + beta_2*SEX + epsilon`.

This gives us the 95% confidence interval [0.600, 0.902]. We can see that this has changed from the confidence interval from the previously fitted linear model which was [0.596, 0.905]. However, the corresponding p-value for the education has not changed since it remains to be 0. The confidence intervals remain to be relatively similar before creating the additive model which indicates that there is not a strong relationship between education and sex. If there was a strong relationship between these two, then the confifdnce interval would have changed more significantly.

# Part B

The slope parameter beta_1 associated with education in this case adjusts for sex. Previously (in activity #2), we used education to predict wage which didn't account for sex. However we've now split apart the data for male and females so beta_1 now represents the relationship between education and wage when sex is held constant. What this means is that if we look at only the males we observe that an increase in education by 1 is associated with an increase by 0.751 in wage. Similarly, we would expect the same pattern for females. However, we now have a term for sex so just because the slopes are the same for education does not necessarily mean that the final predicted wage would be the same for a male or female with the same education level since there is now an additional slope term for sex.

# Part C

```r
wage_model <- lm(wage ~ factor(occupation), data = wage_data)
get_regression_table(wage_model)
```

When fitting a linear model with occupation as the sole predictor variable and wage as the reponse variable, we need to account for the fact that occupation is a categorical variable rather than a continuous numerical variable. Using factor(), R will create dummy variables allowing us to separate out the different occupations and calculate slope for each one. In the table above, we can see that occupations 2, 3, 4, and 6 vary by a relatively significant amount compared to the reference level which in this case would be occupation 1. We can take a closer look using ANOVA.

```r
anova(wage_model)
```

The p-value in the table above indicates that wage for each occupation vary significantly from one another since the p value is 4.12*10^-21 which is essentially 0 and is therefore lower than 0.05. Therefore, there are significant wage differences among different occupations (for at least one of the occupations compared to the rest).
